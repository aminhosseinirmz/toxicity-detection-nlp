```python
device = torch.device("cpu")
if torch.cuda.is_available():
    device = torch.device("cuda:0")

device
```

```python
# Load Datasets And Create Data Loaders

train_df = pd.read_csv("data/train.csv")
train_df.describe()
```

```python
BATCH_SIZE = 256
# df_frac = 0.3
df_frac = 1.0
num_epochs = 50
```

## MLP (50 tokens - Embedding 300 - adam)

```python
train_dl_50t_embedding_300, val_dl_50t_embedding_300 = get_dataloaders(
    target_convertor="vocab",
    maximum_tokens=50,
    batch_size=BATCH_SIZE,
    dataset_fraction=df_frac,
)
```

```python
vocab_size = 229158
```

```python
mlp = MLP(
    embed_dim=300,
    vocab_size=vocab_size,
    optimizer_type="adam",
    learning_rate=0.001,
    maximum_tokens=50,
).to(device)

mlp.train_model(
    train_dataloader=train_dl_50t_embedding_300,
    validation_dataloader=val_dl_50t_embedding_300,
    num_of_epochs=num_epochs,
    device=device,
    name="MLP (50 tokens - Embedding_300 - adam)",
    architecture="MLP",
)
```
